```python
# Import our tools
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix

print("All libraries loaded successfully!")
```

```python
df = sns.load_dataset( 'penguins')
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# missing 2 values in bill_length_mm, bill_depth_mm , flipper_length_mm, body_mass_g, sex

df['bill_length_mm'].fillna(df['bill_length_mm'].mode()[0],inplace = True)
df['bill_depth_mm'].fillna(df['bill_depth_mm'].mode()[0],inplace = True)
df['flipper_length_mm'].fillna(df['flipper_length_mm'].mode()[0],inplace = True)
df['body_mass_g'].fillna(df['body_mass_g'].mode()[0],inplace = True)
df['sex'].fillna(df['sex'].mode()[0],inplace = True)
```

```python
df.isnull().sum()
```

```python
df
```

```python
df['sex'].unique()
```

```python
# changing strings into numbers

df['species'] = df['species'].map({'Adelie':1, 'Chinstrap':2, 'Gentoo':3})
df['island'] = df['island'].map({'Torgersen':1, 'Biscoe':2, 'Dream':3})
df['sex'] = df['sex'].map({'Male':1, 'Female':2})
```

```python
df
```

```python
# x = except species (our features)
# y = species (what we want to predict)

x = df.drop(columns=['species'])
y = df['species']

print('X shape:', x.shape)
print('Y shape:', y.shape)
```

```python
# split data into train and test

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.4, random_state=29)

print('Training set:', x_train.shape)
print('Test set:', x_test.shape)
```

```python
# create and train the model

model = LogisticRegression(max_iter=1000)
model.fit(x_train, y_train)

print('Model trained successfully')
```

```python
# make predictions on test data

y_pred = model.predict(x_test)

# check accuracy

accuracy = accuracy_score(y_test, y_pred)
print(f'accuracy:{accuracy * 100:.2f}%')

# confusion matrix
cm = confusion_matrix(y_test, y_pred)
print('\nConfusion matrix')
print(cm)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
# train the model

model_rf = RandomForestClassifier(n_estimators=100, random_state=40)
model_rf.fit(x_train, y_train)

print('Model trained successfully')
```

```python
# make predictions

y_pred_rf = model_rf.predict(x_test)

# check accuracy

accuracy_rf = accuracy_score(y_test, y_pred_rf)
print(f'Logistic Regression Accuracy:97.10%')
print(f'Random Forest Accuracy:{accuracy_rf * 100:.2f}%')
print(f'\nImprovement:{(accuracy_rf *100) - 97.10:.2f}%')

#  final result

print(classification_report(y_test, y_pred_rf))
```
