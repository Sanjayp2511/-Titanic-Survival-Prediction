```python
# Import our tools
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix

print("All libraries loaded successfully!")
```

```python
sns.get_dataset_names()
```

```python
df = sns.load_dataset('iris')
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['species'].unique()
```

```python
df['species']=df['species'].map({'setosa':1, 'versicolor':2, 'virginica':3})
df
```

```python
x = df.drop(columns=['species'])
y = df['species']

print('x shape:',x.shape)
print('y shape:',y.shape)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=10)

print('Training set:', x_train.shape)
print('Test set:', x_test.shape)
```

```python
model = LogisticRegression(max_iter=100)
model.fit(x_train, y_train)
print('model trained successfully')
```

```python
y_pred  = model.predict(x_test)

accuracy = accuracy_score(y_test, y_pred)
print(f'accuracy:{accuracy * 100:.2f}%')

cm = confusion_matrix(y_test, y_pred)
print('\n confusion matrix')
print(cm)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier

model_rf =  RandomForestClassifier(n_estimators=5, random_state=20)
model_rf.fit(x_train, y_train)
print('model trained successfully')
```

```python
y_pred_rf = model_rf.predict(x_test)
accuracy_rf= accuracy_score(y_test, y_pred_rf)
print(f'Random Forest Accuracy:{accuracy_rf * 100:.2f}%')

# Full report
print(classification_report(y_test, y_pred_rf))
```

```python
from sklearn.tree import DecisionTreeClassifier

model_dt = DecisionTreeClassifier(random_state=42)
model_dt.fit(x_train, y_train)
y_pred_dt = model_dt.predict(x_test)

print("DECISION TREE")
print(f"Accuracy: {accuracy_score(y_test, y_pred_dt)*100:.2f}%")
print(classification_report(y_test, y_pred_dt))
```

```python
from sklearn.neighbors import KNeighborsClassifier

model_knn = KNeighborsClassifier(n_neighbors=3)
model_knn.fit(x_train, y_train)
y_pred_knn = model_knn.predict(x_test)

print("KNN")
print(f"Accuracy: {accuracy_score(y_test, y_pred_knn)*100:.2f}%")
print(classification_report(y_test, y_pred_knn))
```

```python
from sklearn.svm import SVC

model_svm = SVC(kernel='rbf', random_state=42)
model_svm.fit(x_train, y_train)
y_pred_svm = model_svm.predict(x_test)

print("SVM")
print(f"Accuracy: {accuracy_score(y_test, y_pred_svm)*100:.2f}%")
print(classification_report(y_test, y_pred_svm))
```

```python
from sklearn.naive_bayes import GaussianNB

model_nb = GaussianNB()
model_nb.fit(x_train, y_train)
y_pred_nb = model_nb.predict(x_test)

print("NAIVE BAYES")
print(f"Accuracy: {accuracy_score(y_test, y_pred_nb)*100:.2f}%")
print(classification_report(y_test, y_pred_nb))
```

```python
from xgboost import XGBClassifier
from sklearn.preprocessing import LabelEncoder

# XGBoost needs numbers not text for target
le = LabelEncoder()
y_train_encoded = le.fit_transform(y_train)
y_test_encoded = le.transform(y_test)

model_xgb = XGBClassifier(random_state=42)
model_xgb.fit(x_train, y_train_encoded)
y_pred_xgb = model_xgb.predict(x_test)

print("XGBOOST")
print(f"Accuracy: {accuracy_score(y_test_encoded, y_pred_xgb)*100:.2f}%")
```

```python
results = {
    'Logistic Regression': accuracy_score(y_test, y_pred),
    'Decision Tree': accuracy_score(y_test, y_pred_dt),
    'Random Forest': accuracy_score(y_test, y_pred_rf),
    'KNN': accuracy_score(y_test, y_pred_knn),
    'SVM': accuracy_score(y_test, y_pred_svm),
    'Naive Bayes': accuracy_score(y_test, y_pred_nb),
    'XGBoost': accuracy_score(y_test_encoded, y_pred_xgb)
}

print("\nMODEL COMPARISON")
print("-" * 35)
for model, acc in sorted(results.items(), 
                          key=lambda x: x[1], 
                          reverse=True):
    print(f"{model:25} {acc*100:.2f}%")
```
