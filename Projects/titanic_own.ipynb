```python
# Import our tools
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix

print("All libraries loaded successfully!")
```

```python
dp=sns.load_dataset('titanic')
dp
```

```python
dp.isnull().sum()
```

```python
dp['age'].fillna(dp['age'].median(), inplace = True)

dp['embarked'].fillna(dp['embarked'].mode()[0], inplace = True)

dp['embark_town'].fillna(dp['embark_town'].mode()[0], inplace = True)
```

```python
dp.isnull().sum()
```

```python
dp=dp.drop(columns=['embarked','class','deck','alive', 'who', 'adult_male', 'alone'])
dp
```

```python
dp['sex']=dp['sex'].map({'male':1, 'female':2})
# dp['who']=dp['who'].map({'man':1,'woman':2,'child':3})
# dp['adult_male']=dp['adult_male'].map({True:1,False:0})
dp['embark_town']=dp['embark_town'].map({'Southampton':1,'Cherbourg':2,'Queenstown':3})
# dp['alone']=dp['alone'].map({False:0,True:1})
dp.head()
```

```python
x = dp.drop(columns=['survived'])
y = dp['survived']

print('X shape:',x.shape)
print('Y shape:',y.shape)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=22)

print('training set:', x_train.shape)
print('test set:', x_test.shape)
```

```python
model = LogisticRegression(max_iter=1000)
model.fit(x_train, y_train)

print('model trained successfully')
```

```python
y_pred = model.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)

print(f'accuracy:{accuracy * 100:.2f}%')

cm = confusion_matrix(y_test, y_pred)
print('Confusion Matrix')
print(cm)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier

# Create Random Forest model
# n_estimators = number of trees in the forest
model_rf = RandomForestClassifier(n_estimators=100, random_state=42)

# Train
model_rf.fit(x_train, y_train)

print("Random Forest trained successfully!")
```

```python
# Make predictions
y_pred_rf = model_rf.predict(x_test)

# Evaluate
accuracy_rf = accuracy_score(y_test, y_pred_rf)
print(f"Logistic Regression Accuracy: 77.65%")
print(f"Random Forest Accuracy: {accuracy_rf * 100:.2f}%")
print(f"\nImprovement: {(accuracy_rf * 100) - 77.65:.2f}%")

# Full report
print(classification_report(y_test, y_pred_rf))
```
