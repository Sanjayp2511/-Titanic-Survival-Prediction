```python
print('Welcome to Number guessing game !')
print('Guess the number between 1 to 100')

secret_number = 98

for i in range(1,8):
    user_input = int(input('Enter your number:'))
    if user_input < 1 or user_input > 100:
        print('Invalid number. Please enter the valid number')
    elif user_input > secret_number:
         print('Your number is higher')
    elif user_input < secret_number:
          print('Your number is lower')
    else:
          print('Your guessed number is correct')
          break
if(secret_number == user_input):
    print(f'Congradulataions, You have guessed the number {secret_number} in {i} attempts')
else:
    print('Sorry! All of your trails are exceeded')
```
