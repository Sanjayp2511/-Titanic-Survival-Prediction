```python
user_data = {'Name':'Sanjay','Role':'Team_member'}

def login(user_data):
    print(f"{user_data['Name']} has logged in")

def view_task(user_data):
    print(f"{user_data['Name']} has viewed the tasks")  

    
def assign_task(user_data):
    if(user_data['Role']=='Manager'):
        print(f"{user_data['Name']} has assigned the tasks")
    else:
        print("The team member can't assign the task ")

def submit_task(user_data):
     if(user_data['Role']=='Team_member'):
        print(f"{user_data['Name']} has submitted the task")
     else:
        print("The manager no need to submit the task ")

          
    



login(user_data)
view_task(user_data)
assign_task(user_data)
submit_task(user_data)
```
